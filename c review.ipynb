## Tipos

Algunos de los tipos más importantes en C y su espacio en memoria:

```c
#include "stdio.h"

int main(){
    printf("\nType    bytes");
    printf("\nchar:   %li", sizeof(char));
    printf("\nshort:  %li", sizeof(short));
    printf("\nint:    %li", sizeof(int));
    printf("\nlong:   %li", sizeof(long));
    printf("\nfloat:  %li", sizeof(float));
    printf("\ndouble: %li", sizeof(double));
    return 0;
}
```

Tamaño ocupado en memoria por los apuntadores:

```c
#include "stdio.h"

int main(){
    char *a;
    short *b;
    int *c;
    long *d;
    float *e;
    double *f;
    printf("\nType    bytes");
    printf("\nchar:   %li", sizeof(a));
    printf("\nshort:  %li", sizeof(b));
    printf("\nint:    %li", sizeof(c));
    printf("\nlong:   %li", sizeof(d));
    printf("\nfloat:  %li", sizeof(e));
    printf("\ndouble: %li", sizeof(f));
    return 0;
}
```

## Pointers


```c
#include "stdio.h"
#define NUMDATA 100

int print(int *data){
    int i;
    for(i = 0; i < NUMDATA; i++){
        printf("%i ", data[i]);
    }
    return 0;
}

int main(){
    int i, a[NUMDATA];
    
    for(i = 0; i < NUMDATA; i++){
        a[i] = i;
    }
    
    print(a);
    return 0;
}
```

Uso de memoria dinámica:

```c
#include "stdio.h"
#include "stdlib.h"

#define NUMDATA 100

int print(int *data){
    int i;
    for(i = 0; i < NUMDATA; i++){
        printf("%i ", data[i]);
    }
    return 0;
}

int main(){
    int i, *a;
    a = (int *)malloc(NUMDATA * sizeof(int));
    if(a == NULL){
        perror("malloc error:");
    }
    
    for(i = 0; i < NUMDATA; i++){
        a[i] = i;
    }
    
    print(a);
    free(a);
    return 0;
}
```

## Archivos

```c
#include <stdio.h>
#include <stdlib.h>

#define LEN 10

int main(int argc, char *argv[])
{
    FILE *ap;
    int r;
    
    ap = fopen("test.dat", "w");
    if(ap <= 0){
        perror("Error in fopen: ");
    }
    r = fwrite("hola mundo", sizeof(char), LEN, ap);
    if(r <= 0){
        perror("Error in fwrite: ");
    }     
    fclose(ap);
}
```

```c

#include <stdio.h>
#include <stdlib.h>

#define LEN 10

int main(int argc, char *argv[])
{
    FILE *ap;
    int r;
    char buffer[32];
    ap = fopen("test.dat", "r");
    if(ap <= 0){
        perror("Error in fopen: ");
    }
    r = fread(buffer,sizeof(char), LEN, ap);
    if(r <= 0){
        perror("Error in fread: ");
    }    
    buffer[r] = 0;
    printf("\n%s", buffer);
    fclose(ap);
}
```

```c
#include "stdio.h"
#include "stdlib.h"

struct dataType{
    char nombre[32];
    char tipo[32];
    int edad;
    char raza[16];
    int estatura;
    float peso;
    char sexo;
};


int loadStruct(void *ap){
    struct dataType *data; 
    data = ap;
    //scanf(" %s", data->nombre); 
    sprintf(data->nombre,"Ringo");
    //scanf(" %s", data->tipo);
    sprintf(data->tipo,"perro");
    //scanf(" %i", &data->edad);
    data->edad = 10;
    //scanf(" %s", data->raza);
    sprintf(data->raza,"westie");
    //scanf(" %i", &data->estatura);
    data->estatura = 30;
    //scanf(" %f", &data->peso);
    data->peso = 10.1;
    //scanf(" %c", &data->sexo);
    data->sexo = 'M';
    return 0;
}

int printStruct(void *ap){
    struct dataType *data; 
    data = ap;
    printf("\n nombre: %s", data->nombre);
    printf("\n tipo: %s", data->tipo);
    printf("\n edad: %i", data->edad);
    printf("\n raza: %s", data->raza);
    printf("\n estatura: %i", data->estatura);
    printf("\n peso: %f", data->peso);
    printf("\n sexo: %c", data->sexo);
}


int guardar(void *ap){
    struct dataType *data;
    data = ap;
    
    FILE *apFile;
    int r;
    
    apFile = fopen("datos.dat","w+");
    if(apFile == NULL){
        perror("error fopen:");
        exit(-1);
    }
    
    r = fwrite(ap, sizeof(struct dataType), 1, apFile);
    if(r <= 0){
        perror("error fwrite");
        exit(-1);
    }
    
    r = fclose(apFile);
    if(r < 0){
        perror("error fclose: ");
        exit(-1);
    }    
    
}

int main(){
    struct dataType *data;
    data = malloc(sizeof(struct dataType));
    if(data == NULL){
        perror("error malloc: ");
        exit(-1);
    }
    loadStruct((void *)data);
    printStruct((void *)data);
    guardar((void *)data);
    free(data);
}
```
